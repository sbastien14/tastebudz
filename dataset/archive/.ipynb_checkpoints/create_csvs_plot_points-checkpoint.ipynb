```python
#maps for projection
world_map = f.Map(location=[76.1337, 43.0387], tiles="OpenStreetMap", zoom_start=9)
```

```python
def create_geodf(my_df):
    my_gdf = gp.GeoDataFrame(my_df, geometry=gp.points_from_xy(my_df.lng, my_df.lat))
    return my_gdf
```

```python
def plot_geofence(my_gpf, plot_map):
    #iterate through coordinates and put them on map
    for row in my_gpf.itertuples(index=False):
        lat = row.lat
        lng = row.lng
        folium.Circle(location=[lng, lat], radius=50000).add_to(plot_map)  
    return plot_map
```

```python
test_gpf = create_geodf(test_data)
#display(test_gpf)
type(test_data.lat)
type(test_data.lng)
plot_geofence(test_gpf, world_map)
```
